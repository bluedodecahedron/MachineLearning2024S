```python
import numpy as np
import openml
import sys
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(threshold=sys.maxsize)
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Get dataset by name
dataset = openml.datasets.get_dataset('eucalyptus', download_data=True, download_qualities=True, download_features_meta_data=True)

# Get the data itself as a dataframe (or otherwise)
X, y, _, _ = dataset.get_data(dataset_format="dataframe")
```

```python
X
```

```python
vis_features = ['Year', 'Surv', 'Rainfall', 'Frosts', 'Altitude', 'Crown_Fm', 'Ht', 'Stem_Fm', 'PMCno', 'Utility', 'Vig', 'Ins_res']
plt.rcParams['figure.figsize'] = [15, 10]
rows = int(np.ceil(len(vis_features)/3))
fig, axs = plt.subplots(rows, 3)

for i, f in enumerate(vis_features):
    plt.subplots_adjust(wspace=0.4, hspace=0.6)
    x = X[f]
    m = i // 3
    n = i % 3

    axs[m,n].hist(x, bins=15, color='skyblue', edgecolor='black')
    axs[m,n].set_xlabel(f)
    axs[m,n].set_ylabel('Frequency')

plt.show()
```

```python
# Get dataset by name
dataset = openml.datasets.get_dataset('Stock-Information', download_data=True, download_qualities=True, download_features_meta_data=True)

# Get the data itself as a dataframe (or otherwise)
X, y, _, _ = dataset.get_data(dataset_format="dataframe")
```

```python
X
```

```python
vis_features = ['margin_initial_ratio', 'maintenance_ratio', 'country', 'day_trade_ratio', 'type', 'default_collar_fraction', 'open', 'high', 'low', 'volume', 'average_volume_2_weeks', 'dividend_yield']
plt.rcParams['figure.figsize'] = [15, 10]
rows = int(np.ceil(len(vis_features)/3))
fig, axs = plt.subplots(rows, 3)

for i, f in enumerate(vis_features):
    plt.subplots_adjust(wspace=0.4, hspace=0.6)

    x = X[f]
    # Check if it's a string or a number
    if isinstance(x[0], str):
        # Initialize label encoder
        label_encoder = LabelEncoder()
        # Fit and transform the string array
        x = label_encoder.fit_transform(x)

    x = x[~np.isnan(x)]
    x = (x - x.mean())/x.std()

    # filter outliers
    Q1 = np.percentile(x, 10)-0.01
    Q3 = np.percentile(x, 90)+0.01
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    x = x[(x > lower_bound) & (x < upper_bound)]

    m = i // 3
    n = i % 3

    axs[m,n].hist(x, bins=15, color='skyblue', edgecolor='black')
    axs[m,n].set_xlabel(f)
    axs[m,n].set_ylabel('Frequency')

plt.show()
```
